# Code Javascript để lấy data các link bài báo và label của nó

## đường dẫn lấy data: https://www.politifact.com/factchecks/list/

```python
// Function to extract data and create CSV
function extractAndCreateCSV() {
    // Selecting all the list items
    var listItems = document.querySelectorAll('li.o-listicle__item');

    // Creating a CSV string with headers
    var csv = 'url,label\n';

    // Loop through each list item
    listItems.forEach(function(item) {
        // Extract URL
        var url = item.querySelector('.m-statement__quote-wrap a').getAttribute('href');

        // Extract label
        var label = item.querySelector('.m-statement__meter img').getAttribute('alt');

        // Append data to CSV string
        csv += '"' + url + '","' + label + '"\n';
    });

    // Create a Blob object with the CSV data
    var blob = new Blob([csv], { type: 'text/csv' });

    // Create a temporary anchor element
    var link = document.createElement('a');
    link.href = window.URL.createObjectURL(blob);
    link.download = 'data.csv';

    // Trigger the download
    link.click();
}

// Call the function to extract data and create CSV
extractAndCreateCSV();
```

```python
import csv
import requests
from bs4 import BeautifulSoup

# Function to scrape data from a URL
def scrape_data(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.text, "html.parser")
    article = soup.find("article", class_="m-statement")
    speaker = article.find("a", class_="m-statement__name").text.strip()
    venue = article.find("div", class_="m-statement__desc").text.strip()
    statement = article.find("div", class_="m-statement__quote").text.strip()
    categories = [tag.text.strip() for tag in article.find_all("a", class_="c-tag")]
    return speaker, venue, categories, statement

# Read URLs and labels from the CSV file
input_csv_file = "data_everyone_full.csv"
output_csv_file = "liar_everyone_full_2018_2024.csv"

with open(input_csv_file, mode='r') as infile, open(output_csv_file, mode='w', newline='') as outfile:
    reader = csv.reader(infile)
    writer = csv.writer(outfile)
    writer.writerow(["url", "label", "speaker", "venue", "categories", "statement"])  # Write header
    next(reader)  # Skip header row in input CSV
    for row in reader:
        url = "https://www.politifact.com" + row[0]
        label = row[1]
        try:
            speaker, venue, categories, statement = scrape_data(url)
            venue = f'stated on {venue}'
            categories_str = ','.join(categories)
            writer.writerow([url, label, speaker, venue, categories_str, statement])
        except Exception as e:
            print(f"Error processing URL: {url}")
            print(e)
```

```python

# Load your CSV file into a DataFrame
df = pd.read_csv('liar_everyone_full_2018_2024.csv')

# Define a function to remove the speaker's name and uppercase text
def clean_text(row):
    # Get the speaker's name
    speaker_name = row['speaker']
    # Remove the speaker's name and uppercase the text
    cleaned_text = row['categories'].replace(speaker_name, "").lower().rstrip(', ')
    return cleaned_text

# Apply the function to the DataFrame
df['categories'] = df.apply(clean_text, axis=1)

# Save the updated DataFrame to a new CSV file
df.to_csv('liar_everyone_full_2018_2024.csv', index=False)
```

```python
import pandas as pd
import re

# Load your CSV file into a DataFrame
df = pd.read_csv('liar_everyone_full_2018_2024.csv')

# Define a function to replace spaces with hyphens, convert to lowercase, and remove the last comma
def clean_text(text):
    # Replace spaces with hyphens and convert to lowercase
    cleaned_text = text.replace(" ", "-").lower()
    # Remove trailing commas (if needed)
    cleaned_text = re.sub(r',+$', '', cleaned_text)
    return cleaned_text

# Apply the function to the specified column "subject"
df['speaker'] = df['speaker'].apply(clean_text)

# Save the updated DataFrame to a new CSV file
df.to_csv('liar_everyone_full_2018_2024.csv', index=False)
```

```python
# Load the old dataset
old_df = pd.read_csv('Liar_Dataset.csv')

# Create a new DataFrame with the same structure but new values
new_df = pd.read_csv('liar_everyone_full_2018_2024.csv')
new_columns = ["speaker's job title", 'state info', 'party affiliation', 'barely true counts', 'false counts', 'half true counts', 'mostly true counts', 'pants on fire counts']
# Define a function to map and replace values from old_df to new_df
def replace_column_values(new_df, old_df, column_name, key_column='speaker'):
    # Create a mapping from old_df
    mapping = old_df.set_index(key_column)[column_name].to_dict()
    # Replace values in new_df based on the mapping
    new_df[column_name] = new_df[key_column].map(mapping)

# Replace column values in the new dataset
for col in new_columns:
    replace_column_values(new_df, old_df, col)

new_df.to_csv('liar_everyone_full_2018_2024_v2.0.csv', index=False)
```
