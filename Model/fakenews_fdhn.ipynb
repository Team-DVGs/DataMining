## **Dự đoán với mô hình 1.1.TextCNN**

```python
import sys
sys.path.append('/content/Python-3.10.0a1/FDHN/model')

import pandas as pd
import numpy as np

import torch
import torch.nn as nn
import torch.nn.functional as F

from TC_model import textProcess, LiarDataset, LiarModel
from torch.utils.data import DataLoader
import torch.utils.data as data

# tiền xử lý dữ liệu

num_epochs = 10
batch_size = 8
learning_rate = 1e-3
num_classes = 6
padding_idx = 0
metadata_each_dim = 10

vocab_size = 30522
embedding_dim = 32
n_filters = 32
filter_sizes = [2,3,4]
output_dim = 6
dropout = 0.5
padding_idx = 0
input_dim = 6 * metadata_each_dim + 5
hidden_dim = 64
n_layers = 2
bidirectional = True

col = ['id', 'label', 'statement', 'subject', 'speaker', 'job_title', 'state_info', 'party_affiliation', 'barely_true_counts', 'false_counts', 'half_true_counts', 'mostly_true_counts', 'pants_on_fire_counts', 'context']

label_map = {0: 'pants-fire', 1: 'false', 2: 'barely-true', 3: 'half-true', 4: 'mostly-true', 5: 'true'}
label_convert = {'pants-fire': 0, 'false': 1, 'barely-true': 2, 'half-true': 3, 'mostly-true': 4, 'true':5}

test_data = pd.read_csv('/content/Python-3.10.0a1/FDHN/LIAR/predict.tsv', sep = '\t', names = col)

# print(test_data.head(5))

test_data['label'] = test_data['label'].replace(label_convert)

test_text = torch.tensor(textProcess(test_data['statement'].tolist())['input_ids'])
test_label = torch.nn.functional.one_hot(torch.tensor(test_data['label'].values), num_classes=6).type(torch.float64)
test_subject = torch.tensor(textProcess(test_data['subject'].tolist(), metadata_each_dim)['input_ids'])
test_speaker = torch.tensor(textProcess(test_data['speaker'].tolist(), metadata_each_dim)['input_ids'])
test_job_title = torch.tensor(textProcess(test_data['job_title'].tolist(), metadata_each_dim)['input_ids'])
test_state_info = torch.tensor(textProcess(test_data['state_info'].tolist(), metadata_each_dim)['input_ids'])
test_party_affiliation = torch.tensor(textProcess(test_data['party_affiliation'].tolist(), metadata_each_dim)['input_ids'])
test_context = torch.tensor(textProcess(test_data['context'].tolist(), metadata_each_dim)['input_ids'])

test_dataset = LiarDataset(test_data, test_text, test_label, torch.tensor(test_data['label'].replace(label_convert)),
                          test_subject, test_speaker, test_job_title,
                          test_state_info, test_party_affiliation,
                          test_data['barely_true_counts'].tolist(), test_data['false_counts'].tolist(),
                          test_data['half_true_counts'].tolist(), test_data['mostly_true_counts'].tolist(),
                          test_data['pants_on_fire_counts'].tolist(), test_context)
test_loader = data.DataLoader(test_dataset, batch_size=batch_size, shuffle=False)

DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = LiarModel(vocab_size, embedding_dim, n_filters, filter_sizes, output_dim, dropout, padding_idx, input_dim, hidden_dim, n_layers, bidirectional).to(DEVICE)
model.load_state_dict(torch.load('/content/Python-3.10.0a1/FDHN/model/TC.pt'))


class_labels = {
    0: "pants-fire",
    1: "false",
    2: "barely-true",
    3: "half-true",
    4: "mostly-true",
    5: "true"
}

# Sửa đổi hàm predict_class để xử lý dự đoán cho từng mẫu dữ liệu riêng lẻ
def predict_class(model, data_loader):
    predicted_classes = []

    with torch.no_grad():
        for statements, _, _, metadata in data_loader:
            statements = statements.to(DEVICE)
            metadata = metadata.to(DEVICE)

            outputs = model(statements, metadata)
            _, predicted_class = torch.max(outputs, 1)

            # Lặp qua từng dự đoán và lấy tên nhãn từ từ điển ánh xạ
            for pred_class in predicted_class:
                predicted_label = class_labels[pred_class.item()]
                predicted_classes.append(predicted_label)

    return predicted_classes

# Sử dụng hàm để dự đoán và in ra tên nhãn
predicted_classes = predict_class(model, test_loader)
print(predicted_classes)

test_data['predicted_label'] = predicted_classes
result_df = test_data[['label', 'predicted_label']]
result_df.to_csv('/content/Python-3.10.0a1/FDHN/LIAR/TC_predict.csv', index=False)
```
